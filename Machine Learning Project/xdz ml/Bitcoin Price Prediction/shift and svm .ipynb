```python
import numpy as np
from sklearn.model_selection import train_test_split

# Assuming df is your DataFrame and 'Prediction' is the target column
x = df.drop(['Prediction'], axis=1)  # Features
y = df['Prediction']                  # Target variable

# Ensure x and y have the same number of samples
print(f"x shape: {x.shape}, y shape: {y.shape}")

# Split the data
xtrain, xtest, ytrain, ytest = train_test_split(x, y, test_size=0.2, random_state=42)

# Set predictionDays
predictionDays = 30  # Define the number of prediction days

# Ensure the DataFrame has enough rows
if len(df) >= predictionDays:
    predictionDays_array = np.array(df.drop(['Prediction'], axis=1))[-predictionDays:]
    print(predictionDays_array)
else:
    print("The DataFrame does not have enough rows for prediction.")
```

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.svm import SVR

# Assuming df is your DataFrame
# Check for NaN values in 'Prediction'
print("NaN values in target variable (y):", df['Prediction'].isnull().sum())

# Remove rows with NaN in the target variable
df = df.dropna(subset=['Prediction'])

# Define x and y
x = df.drop(['Prediction'], axis=1)  # Features
y = df['Prediction']                  # Target variable

# Check again for NaN values
print("NaN values in target variable (y) after cleaning:", y.isnull().sum())

# Split the data
xtrain, xtest, ytrain, ytest = train_test_split(x, y, test_size=0.2, random_state=42)

# Create and Train the Support Vector Machine (Regression) using radial basis function
svr_rbf = SVR(kernel='rbf', C=1e3, gamma=0.00001)
svr_rbf.fit(xtrain, ytrain)

print("Model training completed successfully.")
```

```python
svr_rbf_confidence = svr_rbf.score(xtest,ytest)
print('SVR_RBF accuracy :',svr_rbf_confidence)
```

```python
svm_prediction = svr_rbf.predict(xtest)
print(svm_prediction)
print()
print(ytest)
```

```python
# Print the model predictions for the next 30 days
svm_prediction = svr_rbf.predict(predictionDays_array)
print(svm_prediction)
print()
#Print the actual price for bitcoin for last 30 days
print(df.tail(predictionDays))
```
