```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data=pd.read_csv("market.csv")
print(data.shape)
```

```python
data.isnull().sum()
```

```python
data.info()
```

```python
print("Dataset contains {} row and {} colums".format(data.shape[0],data.shape[1]))
```

```python
plt.figure(figsize=(14,6))
plt.style.use('fivethirtyeight')
ax= sns.countplot(x='Gender', data=data , palette = 'copper')
ax.set_xlabel(xlabel= "Gender",fontsize=18)
ax.set_ylabel(ylabel = "Gender count", fontsize = 18)
ax.set_title(label = "Gender count in supermarket", fontsize = 20)
plt.show()
```

```python
# Common Aggregation Functions:
# mean(): Computes the mean (average) value.
# sum(): Computes the sum of values.
# min(): Returns the minimum value.
# max(): Returns the maximum value.
# count(): Counts the number of non-null values.
# median(): Returns the median value.
# std(): Computes the standard deviation.
# var(): Computes the variance.
# prod(): Computes the product of all values.
# first(): Returns the first non-null value.
# last(): Returns the last non-null value.
# mode(): Returns the mode (most frequent value).
# nunique(): Counts the number of unique values.
# mad(): Computes the mean absolute deviation.
# quantile(q): Returns values at the given quantile q.
# sem(): Computes the standard error of the mean.
# cumsum(): Computes the cumulative sum.
# cummin(): Computes the cumulative minimum.
# cummax(): Computes the cumulative maximum.
# cumprod(): Computes the cumulative product.
```

```python
data.groupby(['Gender']). agg({'Total':'sum'})
```

```python
data.groupby(['Gender']).agg({'Total': 'mean'}).rename(columns={'Total': 'Average Total'})
```

```python
# Applying multiple aggregations
data.groupby('Gender').agg({'Total': ['mean', 'sum', 'max']})
```

```python
plt.figure(figsize=(14,6)) 
ax = sns.boxplot(x="Branch", y = "Rating" ,data =data, palette= "RdYlBu")
ax.set_title("Rating distribution between branches", fontsize = 25)
ax.set_xlabel(xlabel = "Branches", fontsize = 16)
ax.set_ylabel(ylabel = "Rating distribution", fontsize = 16)
```

```python
data["Time"]= pd.to_datetime(data["Time"])
data["Hour"]= (data["Time"]).dt.hour
plt.figure(figsize=(14,6))
plt.style.use('classic')
SalesTime = sns.lineplot(x="Hour", y ="Quantity", data = data).set_title("product sales per Hour")
```
