```python
import cv2
image = cv2.imread("1.jpg")
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
cv2.imwrite("2.jpg", gray_image)
inverted_image = 255 - gray_image
cv2.imwrite("3.jpg", inverted_image)
blurred = cv2.GaussianBlur(inverted_image, (21, 21), 0)
cv2.imwrite("4.jpg", blurred)
inverted_blurred = 255 - blurred
pencil_sketch = cv2.divide(gray_image, inverted_blurred, scale=250.0)
cv2.imwrite("5.jpg", pencil_sketch)
```

```python
import cv2
import matplotlib.pyplot as plt

# Load the image
image = cv2.imread("5.jpg")

# Convert the image from BGR to RGB (since OpenCV loads images in BGR format)
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

# Display the image using matplotlib
plt.imshow(image_rgb)
plt.axis('off')  # Hide axis
plt.show()
```
